```python
# request again but this time save them in a list so no need to request everytime running the loop 

responses = [] 

for book in books: 
    
    book_link = book.find("a")["href"]
    full_url= base_url + book_link 
    
    detail_response = requests.get(full_url)
    if detail_response.status_code == 200: 
        responses.append(detail_response) 
```

```python
def scrape_books (min_rating, max_price): 

    all_books = []
    rating_map = {"One": 1, "Two": 2, "Three": 3, "Four": 4, "Five": 5}

    for book, response in zip(books, responses): 
        book_info = {}
                       
        book_info["Title"] = book.find("img")["alt"]
        book_info["Rating"] = rating_map[book.find("p", class_="star-rating")["class"][1]]
        book_info["Price"] = float(book.find("p", class_="price_color").get_text().replace("£",""))   


        detail_soup = BeautifulSoup(response.content, "html.parser")

        upc_row = detail_soup.find("th", string = "UPC") 
        book_info["UPC"] = upc_row.find_parent("tr").find("td").get_text()

        avai_row = detail_soup.find("th", string = "Availability") 
        book_info["Availability"] = avai_row.find_parent("tr").find("td").get_text() 

        genre_row = detail_soup.find("ul", class_="breadcrumb") 
        book_info["Genre"] = genre_row.find_all("a")[-1].get_text() 

        description_row = detail_soup.find("div", id="product_description")
        book_info["Description"] = description_row.find_next("p").get_text()

                
        if book_info["Rating"] >= min_rating and book_info["Price"] <= max_price: 
            all_books.append(book_info) 
            
    return pd.DataFrame(all_books)
            

```
